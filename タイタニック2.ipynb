```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline

data = pd.read_csv('titanic_train.csv')

data.isnull().sum()
```

```python
embarked_mapping = {'S':1, 'C':2, 'Q':3}
data['Embarked'] = data['Embarked'].map(embarked_mapping)
data.head()
```

```python
data.dropna(subset=['Embarked'])
#Embarkedに欠損値を含む行を削除
```

```python
sex_mapping = {'male':1, 'female':2}
data['Sex'] = data['Sex'].map(sex_mapping)
data.head()
```

```python
data1=data.fillna(data.Age.mean())
```

```python
data1.isnull().sum()
```

```python
# ロジスティック回帰
data1_pickup = data1.loc[:,['Pclass','Sex','Age','Fare','Survived']]
sns.pairplot(data1_pickup, size=2.0, hue='Survived',markers='+')
plt.show()
```

```python
#相関性が見えないので'Pclass','Sex','Age','Fare'すべてのデータを使用
X = data1.loc[:,['Pclass','Sex','Age','Fare']].values
y = data1.loc[:,['Survived']].values
y = y.reshape(-1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_std = scaler.transform(X)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1.0)
lr.fit(X_train, y_train)
```

```python
print('train acc: %.3f' % lr.score(X_train, y_train))
print('test acc: %.3f' % lr.score(X_test, y_test))
```
