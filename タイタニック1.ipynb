```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline

data = pd.read_csv('titanic_train.csv')
#y = data["survived"].values
```

```python
data.head()
```

```python
embarked_mapping = {'S':1, 'C':2, 'Q':3}
data['Embarked'] = data['Embarked'].map(embarked_mapping)
data.head()
```

```python
data.Embarked.fillna(2.0)
```

```python
sex_mapping = {'male':1, 'female':2}
data['Sex'] = data['Sex'].map(sex_mapping)
data.head()
```

```python
data1=data.fillna(data["Age"].mean())
```

```python
from sklearn.model_selection import train_test_split

X = data1[["Pclass","Fare","Sex","Age","SibSp","Parch","Embarked"]].values
y = data1["Survived"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
data1.isnull().sum()
```

```python
#ランダムフォレスト

from sklearn.ensemble import RandomForestClassifier
rfc_1 = RandomForestClassifier(random_state=0, n_estimators=10)
rfc_1.fit(X_train, y_train)
```

```python
# 正答率を出力
print('train accuracy: %.3f' % rfc_1.score(X_train, y_train))
print('test  accuracy: %.3f' % rfc_1.score(X_test, y_test))
```
